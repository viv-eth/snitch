```python
# import all necessary packages
import pandas as pd
```

```python
# read in the data
mnist_path = "/scratch/msc22f11/msc22f11/snitch/sw/applications/data/mnist_train.csv"
```

```python
mnist_train_df = pd.read_csv(mnist_path)
```

```python
mnist_train_df.head()
```

```python
# get one batch of data
mnist_one_batch_df = mnist_train_df[:256]
mnist_one_batch_df.head()
```

```python
# separate the data and labels
mnist_one_batch_labels = mnist_one_batch_df["label"]
mnist_one_batch_labels.head()
```

```python
mnist_one_batch_images = mnist_one_batch_df.drop(["label"], axis=1)
# scale the images to [0, 1] by dividing by 255
mnist_one_batch_images_scaled = mnist_one_batch_images.div(255)
mnist_one_batch_images_scaled.head()
```

```python
# pd.set_option('display.max_rows', None)
print(mnist_one_batch_images_scaled.loc[0])
```

```python
# functions to convert a float to hex
from struct import *

def float_to_hex(f):
    hex_val = hex(unpack('<I', pack('<f', f))[0])
    return hex_val.upper()
def double_to_hex(f):
    return hex(unpack('<Q', pack('<d', f))[0])
```

```python
# convert the labels to hex
mnist_one_batch_labels_hex = mnist_one_batch_labels.apply(lambda x: "0x" + hex(x)[2:].zfill(8))
mnist_one_batch_labels_hex
```

```python
# convert the images to hex
mnist_one_batch_images_hex = mnist_one_batch_images_scaled.applymap(lambda x: "0x" + float_to_hex(x)[2:].zfill(8))
mnist_one_batch_images_hex.head()
```

```python
pd.set_option('display.max_rows', None)
print(mnist_one_batch_images_hex.loc[0])
```

```python
# generate the dram preload file for the batch of images
start_address = "@80040000\n" #80000000 is the start address of the dram according to config
print("address before inserting all elements: 0x{}".format(start_address[1:])) 
hex_addr = 0x80040000
f = open("/scratch/msc22f11/msc22f11/snitch/sw/applications/data/dram_preload_mnist_images_epoch.txt", "w")
f.write(start_address)
# each row corresponds to an image
for index, row in mnist_one_batch_images_hex.iterrows():
    # each column corresponds to the respective pixel in the selected row
    for col in range(0, len(mnist_one_batch_images_hex.columns)):
        f.write(str(row[col])+"\n")
        hex_addr += 0x4
f.close() 
print("address after inserting all elements: 0x{:X}".format(hex_addr))   
```

```python
# generate the dram preload file for the batch of images from the UNHEXED DF
start_address = "@80040000\n" #80000000 is the start address of the dram according to config
print("address before inserting all elements: 0x{}".format(start_address[1:])) 
hex_addr = 0x80040000
f = open("/scratch/msc22f11/msc22f11/snitch/sw/applications/data/dram_preload_mnist_images_epoch_nohex.txt", "w")
f.write(start_address)
# each row corresponds to an image
for index, row in mnist_one_batch_images_scaled.iterrows():
    # each column corresponds to the respective pixel in the selected row
    for col in range(0, len(mnist_one_batch_images_scaled.columns)):
        f.write(str(row[col])+"\n")
        hex_addr += 0x4
f.close() 
print("address after inserting all elements: 0x{:X}".format(hex_addr))   
```

```python
# generate the dram preload file for the batch of labels
start_address = "@80108000\n"
print("address before inserting all elements: 0x{}".format(start_address[1:])) 
hex_addr = 0x80108000
f = open("/scratch/msc22f11/msc22f11/snitch/sw/applications/data/dram_preload_mnist_labels_epoch.txt", "w")
f.write(start_address)
# each row corresponds to an image
for row in mnist_one_batch_labels_hex:
    f.write(str(row)+"\n")
    hex_addr += 0x4
f.close() 
print("address after inserting all elements: 0x{:X}".format(hex_addr))      
```
