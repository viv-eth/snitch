```python
from tkinter.tix import ButtonBox
from turtle import color
import matplotlib.pyplot as plt
import numpy as np

NUM_MAX_TOT_CYCLES = [6734, 6738, 6661, 6692, 6640, 6615, 6566, 6511]

print(np.max(NUM_MAX_TOT_CYCLES))

NUM_COMPUTE_CYCLES = [3132*2] * 8 # number of fmadd operations

NUM_OVERHEAD_CYCLES = [x - y for x, y in zip(NUM_MAX_TOT_CYCLES, NUM_COMPUTE_CYCLES)]

plt.rcParams["figure.figsize"] = (10,8)
# plot the overhead cycles and compute cycles in a bar graph
plt.bar(range(len(NUM_COMPUTE_CYCLES)), NUM_MAX_TOT_CYCLES, color='#ffa600', label='Overhead Cycles')
plt.bar(range(len(NUM_COMPUTE_CYCLES)), NUM_COMPUTE_CYCLES, color='#003f5c', label='Compute Cycles')
# write the number of overhead cycles in the bar graph
for i, v in enumerate(NUM_OVERHEAD_CYCLES):
    plt.text(i, NUM_COMPUTE_CYCLES[i] + 50, str(v), color='black', ha='center', va='bottom')
for i, v in enumerate(NUM_COMPUTE_CYCLES):
    plt.text(i, v - 300, str(v), color='white', ha='center', va='bottom')
# plot legend next to the bar graph
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Core ID')
plt.ylabel('Cycles')
plt.title('Overhead and Compute Cycles FeedForward Opt Inline')
plt.show()
```
